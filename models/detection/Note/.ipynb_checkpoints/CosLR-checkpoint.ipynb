```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
init_lr = 0.01
total_epoch = 30

x = np.arange(0, 30, 1)
y = (0.5 * (1. + np.cos(x/total_epoch * np.pi))) * init_lr

z = np.zeros_like(y)
z[:int(len(z)*0.7)] = init_lr
z[int(len(z)*0.7):int(len(z)*0.9)] = init_lr * 0.1
z[int(len(z)*0.9):] = init_lr * 0.01

fig, ax = plt.subplots()  # Create a figure and an axes.
ax.plot(x, y, label='coslr')  # Plot some data on the axes.
ax.plot(x, z, label='steplr')  # Plot more data on the axes...
ax.set_xlabel('Epoch')  # Add an x-label to the axes.
ax.set_ylabel('Learning Rate')  # Add a y-label to the axes.
ax.set_title("LR Scheduler")  # Add a title to the axes.
ax.legend()  # Add a legend.
```
